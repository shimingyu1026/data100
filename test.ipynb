```python
import matplotlib.pyplot as plt



fig, ax = plt.subplots(dpi=600,figsize=(4, 4))

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

#width: 柱子变细
ax.bar(fruits, counts, label=bar_labels, color=bar_colors, edgecolor='black', linewidth=2.5,width=0.618,)

ax.set_ylabel('fruit supply', fontsize=12, fontweight='bold')
ax.set_title('Fruit supply by kind and color', fontsize=16,fontweight='bold')
ax.legend(title='color', loc='best', fontsize=12, title_fontsize=12)

#设置透明
fig.patch.set_alpha(0.0)
ax.patch.set_alpha(0.0)

#设置显示范围
ax.set_ylim(0, 110)


# 在每个柱子上方添加数值标签
for i, count in enumerate(counts):
    ax.text(i, count + 1, str(count), ha='center', va='bottom', fontsize=12)
fig.tight_layout()
ax.grid(axis='y', linestyle='--', alpha=0.6)


plt.savefig('fruit_bar_chart_high_res.pdf', bbox_inches='tight', transparent=True)
plt.show()

```
